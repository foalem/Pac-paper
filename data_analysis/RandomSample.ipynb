```python
# Load Excel file
df_RQ2 = pd.read_excel("Dataset_PaC_Manual_Analysis_RQ2.xlsx")
df_RQ2
```

```python
def sample_284_rows(df: pd.DataFrame, random_state: int = 42) -> pd.DataFrame:
    """
    Randomly samples 284 rows from the given DataFrame.

    Parameters:
    - df (pd.DataFrame): Input DataFrame to sample from.
    - random_state (int): Seed for reproducibility (default: 42).

    Returns:
    - pd.DataFrame: A new DataFrame containing 284 randomly sampled rows.
    """
    if len(df) < 284:
        raise ValueError(f"DataFrame contains only {len(df)} rows, cannot sample 284.")

    sampled_df = df.sample(n=284, random_state=random_state).reset_index(drop=True)
    return sampled_df
```

```python
sample_284_rows(df_RQ2).to_excel("Dataset_PaC_Manual_Analysis_RQ2_284_Sample.xlsx")
```

```python
df_pac_usage = pd.read_csv("pac_usage_summary_updated.csv")
df_pac_usage
```

```python
df_repos = pd.read_csv("PaC_Repos_final_Dataset.csv")
df_repos
```

```python
def filter_common_full_names(df_pac_usage: pd.DataFrame, df_repos: pd.DataFrame) -> pd.DataFrame:
    """
    Filters df_pac_usage to retain only rows where 'full_name' is also present in df_repos.

    Parameters:
    - df_pac_usage (pd.DataFrame): DataFrame containing PaC tool usage.
    - df_repos (pd.DataFrame): DataFrame containing repository metadata.

    Returns:
    - pd.DataFrame: Filtered df_pac_usage with only common 'full_name' entries.
    """
    common_full_names = set(df_repos['full_name']).intersection(set(df_pac_usage['full_name']))
    return df_pac_usage[df_pac_usage['full_name'].isin(common_full_names)].reset_index(drop=True)
```

```python
filtered_df = filter_common_full_names(df_pac_usage, df_repos)
```
