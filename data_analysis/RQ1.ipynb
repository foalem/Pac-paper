```python
def enrich_usage_with_metadata(df_usage: pd.DataFrame, df_meta: pd.DataFrame) -> pd.DataFrame:
    """
    Enriches df_usage with additional metadata from df_meta based on matching 'full_name'.
    
    Parameters:
    - df_usage (pd.DataFrame): DataFrame containing PaC usage data with a 'full_name' column.
    - df_meta (pd.DataFrame): DataFrame containing metadata to enrich with, must include 'full_name'.
    
    Returns:
    - pd.DataFrame: Enriched DataFrame with metadata columns added.
    """
    # Remove duplicate full_name entries in the metadata to ensure 1-to-1 mapping
    df_meta_unique = df_meta.drop_duplicates(subset="full_name", keep="first")

    # Convert metadata to a lookup dictionary
    meta_dict = df_meta_unique.set_index("full_name").to_dict(orient="index")

    enriched_rows = []

    for _, row in df_usage.iterrows():
        full_name = row["full_name"]
        new_row = row.to_dict()

        # Add matching metadata if available
        if full_name in meta_dict:
            for key, value in meta_dict[full_name].items():
                if key != "full_name":
                    new_row[key] = value
        else:
            # Fill missing metadata with None if not found
            for col in df_meta_unique.columns:
                if col != "full_name":
                    new_row[col] = None

        enriched_rows.append(new_row)

    df_enriched = pd.DataFrame(enriched_rows)
    return df_enriched
```

```python
def count_pac_tool_usage(df: pd.DataFrame) -> pd.DataFrame:
    # Identify PaC tool usage columns (all columns except 'full_name')
    pac_columns = [col for col in df.columns if col != 'full_name']

    # Count how many tools are used per row (non-zero entries)
    df['num_tools_used'] = df[pac_columns].gt(0).sum(axis=1)

    # Define bins including 0 usage
    bins = [-1, 0, 1, 2, 3, float('inf')]
    labels = ['0 tools', '1 tool', '2 tools', '3 tools', '>3 tools']
    df['usage_category'] = pd.cut(df['num_tools_used'], bins=bins, labels=labels, right=True)

    # Aggregate results
    usage_summary = df['usage_category'].value_counts().sort_index().reset_index()
    usage_summary.columns = ['No. PaC Tools Used', 'No. of Projects']

    # Compute percentage
    total = usage_summary['No. of Projects'].sum()
    usage_summary['% No. of Projects'] = (usage_summary['No. of Projects'] / total * 100).round(2)

    return usage_summary
```

```python
def plot_pac_usage_pareto_repos(df: pd.DataFrame, output_path: str = "pac_usage_pareto_Repo.pdf"):
    """
    Plots a Pareto chart based on the percentage of repositories using each PaC tool.
    The bars are labeled with proper PaC tool names instead of internal column names.

    Parameters:
    - df (pd.DataFrame): DataFrame with boolean columns for each PaC tool, prefixed with 'has_'.
    """
    # Mapping from column names to readable tool names
    tool_name_map = {
        "has_rego": "Open Policy Agent (OPA)",
        "has_sentinel": "HashiCorp Sentinel",
        "has_pulumi": "Pulumi",
        "has_cedar": "Cedar Policy Language (CPL)",
        "has_kyverno": "Kyverno OSS",
        "has_custodian": "Cloud Custodian",
        "has_awsconfigcloudgaurd": "AWS Config",
        "has_opagatekeeper": "GateKeeper",
        "has_Kubewarden":"Kubewarden"
    }

    # Filter PaC usage columns and rename them
    pac_columns = [col for col in df.columns if col in tool_name_map]
    df_pac_bool = df[pac_columns].rename(columns=tool_name_map)

    # Calculate usage percentage
    usage_counts = df_pac_bool.sum()
    total_repos = len(df)
    usage_percent = (usage_counts / total_repos * 100).sort_values(ascending=False)
    cumulative_percent = usage_percent.cumsum()

    # Plot
    fig, ax1 = plt.subplots(figsize=(12, 6))
    bars = ax1.bar(usage_percent.index, usage_percent.values, color='gray', edgecolor='black')
    ax1.set_ylabel("No. of Repos (%)", fontsize=16, labelpad=16)
    ax1.set_xlabel("PaC Tools", fontsize=16, labelpad=16)
    # ax1.set_title("Pareto Chart of PaC Tool Usage Across Repositories")
    ax1.set_ylim(0, 100)
    ax1.set_xticks(range(len(usage_percent)))
    ax1.set_xticklabels(usage_percent.index, rotation=25, ha="right", fontsize=14)
    ax1.tick_params(axis='y', labelsize=14)
    ax1.grid(axis='y', linestyle='--', alpha=0.7)

    # Annotate bars
    for bar in bars:
        height = bar.get_height()
        ax1.annotate(f"{height:.0f}%", xy=(bar.get_x() + bar.get_width() / 2, height),
                     xytext=(0, 3), textcoords="offset points", ha='center', va='bottom', fontsize=14)

    # Cumulative percentage line
    ax1.plot(usage_percent.index, cumulative_percent, color='black', marker='o', linewidth=1)

    # Highlight 80% threshold
    threshold_line = 80
    tools_to_reach_threshold = cumulative_percent[cumulative_percent <= threshold_line]
    cutoff_index = len(tools_to_reach_threshold)
    ax1.axhline(y=threshold_line, color='teal', linestyle='--')
    ax1.axvline(x=cutoff_index, color='teal', linestyle='--')
    plt.tight_layout()
    plt.savefig(output_path, format='pdf')
    plt.show()
```

```python
def plot_pac_usage_pareto_single_y(df: pd.DataFrame, output_path: str = "pac_usage_pareto.pdf"):
    """
    Plots a refined Pareto chart for Policy as Code (PaC) tools usage based on file count.
    Enhancements include better axis labels, bar annotations, and clearer text rendering.

    Parameters:
    - df (pd.DataFrame): DataFrame where each column (except 'full_name') represents a PaC tool.
    - output_path (str): File path to save the output PDF chart.
    """
    # Compute totals and percentages
    totals = df.drop(columns=["full_name"]).sum().sort_values(ascending=False)
    total_files = totals.sum()
    percentages = (totals / total_files) * 100
    cumulative = percentages.cumsum()

    threshold = 80
    highlight_index = cumulative[cumulative > threshold].index[0]
    threshold_x_pos = list(percentages.index).index(highlight_index)

    # Create the figure and main axis
    fig, ax1 = plt.subplots(figsize=(12, 6))

    # Bar plot (all in gray)
    bars = ax1.bar(
        percentages.index,
        percentages.values,
        color='gray',
        edgecolor='black'
    )

    # Improve axis labels and title
    ax1.set_ylabel("No. of Files (%)", fontsize=16, labelpad=16)
    ax1.set_xlabel("Policy as Code (PaC) Tools", fontsize=16, labelpad=16)
    # ax1.set_title("Pareto Chart of PaC Tool File Usage", fontsize=14, weight='bold')
    ax1.set_ylim(0, 100)
    ax1.set_xticks(range(len(percentages)))
    ax1.set_xticklabels(percentages.index, rotation=25, ha="right", fontsize=14)
    ax1.tick_params(axis='y', labelsize=14)
    ax1.grid(axis='y', linestyle='--', alpha=0.6)

    # Annotate each bar with percentage
    for bar in bars:
        height = bar.get_height()
        ax1.annotate(f"{height:.0f}%",
                     xy=(bar.get_x() + bar.get_width() / 2, height),
                     xytext=(0, 5), textcoords="offset points",
                     ha='center', va='bottom', fontsize=14)

    # Overlay cumulative percentage line
    ax1.plot(range(len(percentages)), cumulative, color='black', marker='o', linewidth=1)

    # Threshold line for 80% cumulative
    ax1.axhline(threshold, color='teal', linestyle='--', linewidth=1.5)
    ax1.axvline(threshold_x_pos, color='teal', linestyle='--', linewidth=1.5)

    # Final layout and save
    plt.tight_layout()
    plt.savefig(output_path, format='pdf')
    plt.show()
    plt.close()
```

```python
def plot_pac_co_usage_heatmap(df: pd.DataFrame):
    """
    Plots a heatmap showing co-usage percentages between Policy as Code (PaC) tools
    based on their presence in repositories.

    Parameters:
    - df (pd.DataFrame): A DataFrame where each row corresponds to a repository.
                         The 'full_name' column identifies the repository,
                         and remaining columns are PaC tools with usage counts.
    """
    # Drop non-tool column
    tool_df = df.drop(columns=["full_name"])

    # Convert counts to binary presence (1 if any usage, 0 otherwise)
    binary_df = tool_df.gt(0).astype(int)

    # Calculate co-usage matrix
    co_usage_counts = binary_df.T.dot(binary_df)

    # Convert counts to percentages of total repositories
    total_repos = len(df)
    co_usage_percent = (co_usage_counts / total_repos) * 100

    # Plotting the heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(
        co_usage_percent,
        annot=True,
        fmt=".1f",
        cmap="Blues",
        cbar_kws={'label': 'Co-usage (%)'}
    )
    plt.title("PaC Tools Co-usage Heatmap (as % of repositories)")
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
plot_pac_co_usage_heatmap(df_pac_usage)
```

```python
def plot_normalized_co_usage(df: pd.DataFrame, output_path: str = "heatmap_pac_co-usage.pdf"):
    """
    Plots a normalized co-usage heatmap where values represent the percentage of co-usage
    between PaC tools relative to the total usage of each tool (i.e., diagonal is 1.0).
    
    Parameters:
    - df (pd.DataFrame): DataFrame where each column (except 'full_name') is a PaC tool,
                         and each cell indicates file count (non-zero = usage).
    """
    # Prepare binary usage matrix
    tools_only = df.drop(columns=["full_name"])
    binary_matrix = tools_only.gt(0).astype(int)

    # Co-usage count matrix
    co_usage = binary_matrix.T.dot(binary_matrix)

    # Normalize by diagonal to get conditional co-usage
    normalized = co_usage.div(co_usage.values.diagonal(), axis=0)

    # Plot
    plt.figure(figsize=(10, 8))
    ax = sns.heatmap(normalized, annot=True, fmt=".2f", cmap="OrRd", vmin=0, vmax=1)
    # plt.title("Normalized Co-usage Heatmap of PaC Tools")
    ax.set_xlabel("PaC Tools", fontsize=16, labelpad=16)
    ax.set_ylabel("PaC Tools", fontsize=16, labelpad=16)
    plt.xticks(rotation=30, ha='right', fontsize=14)
    plt.yticks(rotation=0, fontsize=14)
    plt.tight_layout()
    plt.savefig(output_path, format='pdf')
    plt.show()
```

```python
def plot_violin_metrics(df: pd.DataFrame, output_path: str = "Dataset_Summary.pdf"):
    """
    Plots a 2x2 grid of violin plots for the following repository metrics:
    - Age (in months)
    - Size
    - Stars
    - Contributors

    Each metric is visualized in a separate subplot with its own y-axis scale.
    A shared legend is shown above the plots.

    Parameters:
    - df (pd.DataFrame): DataFrame containing 'created_at', 'updated_at',
                         'size', 'stargazers_count', 'contributors_count'.
    """
    # Parse date columns and calculate age in months
    df['created_at'] = pd.to_datetime(df['created_at'])
    df['updated_at'] = pd.to_datetime(df['updated_at'])
    df['age_months'] = ((df['updated_at'] - df['created_at']) / pd.Timedelta(days=30)).astype(int)

    # Select and rename the metrics for visualization
    df_viz = df[['age_months', 'size', 'stargazers_count', 'contributors_count']]
    df_viz.columns = ['Age', 'Size', 'Stars', 'Contributors']

    # Melt the dataframe for seaborn plotting
    df_melted = df_viz.melt(var_name="Metric", value_name="Value")

    # Define color palette
    palette = {
        "Age": "#FF9999",
        "Size": "#66B3FF",
        "Stars": "#FFCC99",
        "Contributors": "#99FF99"
    }

    # Create 2x2 grid of subplots
    fig, axes = plt.subplots(2, 2, figsize=(6, 8))
    metrics = ["Age", "Size", "Stars", "Contributors"]
    axes = axes.flatten()

    for i, metric in enumerate(metrics):
        sns.violinplot(data=df_melted[df_melted["Metric"] == metric],
                       x="Metric", y="Value", ax=axes[i],
                       palette={metric: palette[metric]}, inner="point")
        axes[i].set_title(metric)
        axes[i].set_xlabel("")
        axes[i].set_ylabel("value")

    # Add custom legend
    legend_elements = [Patch(facecolor=palette[metric], label=metric) for metric in metrics]
    # fig.legend(handles=legend_elements, title="Metrics", loc='center right')
    plt.tight_layout()
    plt.savefig(output_path, format='pdf')
    plt.show()
```

```python
plot_violin_metrics(df_repos)
```

```python
def save_custom_legend(output_file="legend.pdf"):
    colors = {
        "Age": "#FF9999",
        "Size": "#66B3FF",
        "Stars": "#FFCC99",
        "Contributors": "#99FF99"
    }
    handles = [Patch(color=color, label=label) for label, color in colors.items()]
    
    fig, ax = plt.subplots(figsize=(6, 8))
    ax.axis("off")
    legend = ax.legend(handles=handles, loc='center right', title="Metrics", frameon=False)
    fig.savefig(output_file, bbox_inches='tight')
    plt.close(fig)
```
